```python
# File location and type
file_location = "/FileStore/tables/Book1-3.csv"
file_type = "csv"

df_pyspark=spark.read.csv(file_location,header=True,inferSchema=True)
```

```python
## check the schema
df_pyspark.printSchema()
```

```python
#getting columns
df_pyspark.columns
```

```python
#picking up certain column
df_pyspark.select('Name').show()
```

```python
#picking up multiple columns
df_pyspark.select(['Name','Age']).show()
```

```python
#to describe the data 
df_pyspark.describe().show()
```

```python
# adding columns in dataframe
#df_pyspark.withColumn("column name","value")
df_pyspark.withColumn("age after 2 years",df_pyspark['Age']+2).show()
```

```python
# dropping cloumns
df_pyspark.drop("age after 2 years").show()
```

```python
#how to rename the columns
df_pyspark.withColumnRenamed('Age','age').show()
```

```python
#Handling missing values
#dropping NAN values
df_pyspark.na.drop().show() #if we dont give any values in the drop function the rows will be dropped
```

```python
df_pyspark.na.drop(how="any").show() #if we have any null value in a row then row will be dropped
```

```python
df_pyspark.na.drop(how="all").show() #if all the values in a row are null then row will be dropped
```

```python
#threshold
#checks the non null values according to threshold value
df_pyspark.na.drop(how="any",thresh=2).show() 
```
